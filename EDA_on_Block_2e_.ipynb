### Study Total TB cases over HIV +ve,-ve male and female cases

```python
df.drop(columns = ['Unnamed: 0'],inplace = True)
```

```python
final_num_cols
```

```python
import matplotlib.pyplot as plt

study_HIV_columns = ['Male TB HIV Positive Cases',
                     'Female TB HIV Positive Cases',
                     'Male TB HIV Negative Cases',
                     'Female TB HIV Negative Cases',
                     'Male TB HIV Unknown Cases',
                     'Female TB HIV Unknown Cases']

dfx = df.set_index('LGA')

df_filtered = dfx[study_HIV_columns]

# Group by 'LGA' and sum the values for each group
lga_group = df_filtered.groupby('LGA').sum()

# Calculate the total count for each LGA
lga_group['Total'] = lga_group.sum(axis=1)

# Sort LGAs based on the total count
sorted_lgas = lga_group['Total'].sort_values(ascending=False).index

# Use the sorted order to plot the bars
ax = lga_group[study_HIV_columns].loc[sorted_lgas].plot(kind='bar', stacked=True, figsize=(16, 6), colormap='viridis')

# Add annotations for the total count on top of each stack
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy()
    ax.annotate(f'{height:.0f}', (x + width/2, y + height), ha='center', va='center')

plt.title('HIV TB Counts in Each LGA (Sorted)')
plt.xlabel('LGA')
plt.ylabel('Count')
plt.show()
```

```python
# created a function to plot pie charts
def Pie_chart_plot(column):
  # Check percentage cases of each village by Total of Female TB cases
  check_col = column

  total_sum = df[check_col].sum()
  # get percent of the village
  name = 'Percentage of ' + column
  df[name] = (df[check_col] / total_sum * 100)

  df_aggreageted = df.groupby('LGA')[name].sum().reset_index()
  # sort the df by percentage in ascending order
  dfx = df_aggreageted.sort_values(by = name , ascending = True)
  # dfx=df

  # get colors as per the length
  colors = plt.cm.viridis(np.linspace(0,1,23))

  plt.figure(figsize=(7,7))
  plt.pie(dfx[name], labels = dfx['LGA'],autopct = '%1.1f%%',startangle =90,pctdistance=0.90,colors=colors)
  plt.title(f'{name} in each LGA')
  plt.show()
```

```python
for col in final_num_cols:
  Pie_chart_plot(col)
```
